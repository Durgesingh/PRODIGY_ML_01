```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
axes[0].scatter(data['Living Area'],data['Price'])
axes[0].set_xlabel('Living Area')
axes[0].set_ylabel("Price")
axes[0].set_title("Price vs Living Area")

axes[1].scatter(data['Bathrooms'],data['Price'])
axes[1].set_xlabel('Bathrooms')
axes[1].set_ylabel("Price")
axes[1].set_title("Price vs Bathrooms")

axes[2].scatter(data['Bedrooms'],data['Price'])
axes[2].set_xlabel('Bedrooms')
axes[2].set_ylabel("Price")
axes[2].set_title("Price vs Bedrooms")
plt.show()
```

```python
data.drop(columns = ['Age'],inplace = True)
data.drop(columns = ['Fireplace'],inplace = True)
data.drop(columns = ['Lot Size'],inplace = True)
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['Price'],axis = 1)
Y = data['Price']
```

```python
X
```

```python
Y
```

```python
X_train , X_test , Y_train ,Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(X_train,Y_train)

reg.score(X_test,Y_test)
```
